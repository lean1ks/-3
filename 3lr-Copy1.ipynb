Как различается вероятность женщин и мужчин стать жертвой в зависимости от воз‑раста? Постройте визуализацию. В каком возрастном промежутке мужчины чаще становятся жертвами преступлений?

```python
df1 = df.dropna(subset=['victim_age'])
y=df1.victim_age
y
df2 = df.dropna(subset=['victim_sex'])
x = df2.victim_sex
x
```

```python
fig, ax = plt.subplots(figsize=(9, 7))

bars = plt.hist(x, bins=118, color="purple")
plt.title('Количество жертв мужчин и женщин')
plt.xlabel('Пол')
plt.ylabel('Количество')
plt.grid()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(9, 7))

bars = plt.hist(y, bins=100, color="purple")
plt.title('Возраст жертв')
plt.xlabel('Возраст')
plt.ylabel('Количество')
plt.grid()
plt.show()
```

Люди какого происхождения чаще всего подвергаются преступлениям?

```python
df4 = df.dropna(subset=['victim_descent'])
v = df4.victim_descent
v
```

```python
fig, ax = plt.subplots(figsize=(9, 7))

bars = plt.hist(v, bins=20, color="green")
plt.title('Происхождение жертв')
plt.xlabel('происхождение')
plt.ylabel('Количество')
plt.grid()
plt.show()
```
